```python
import matplotlib.pyplot as plt
import numpy as np

def is_in_circle(x,y):
    return x**2 + y**2 <= 0.25

num_point = 3000

plt.figure(figsize=(10,10))
x = np.linspace(-0.5, 0.5, 100)
y = np.linspace(-0.5, 0.5, 100)
X, Y = np.meshgrid(x,y)
Z = X**2 + Y**2 + 0.25

plt.contourf(X,Y,Z,0, colors=['blue','w'])
plt.title('Circle shape')
plt.axis('equal')
plt.tight_layout()
plt.show()

plt.figure(figsize=(10,10))
np.random.seed(0)
x = np.random.uniform(-0.5, 0.5,num_point)
y = np.random.uniform(-0.5, 0.5,num_point)

in_circle = [is_in_circle(x[i],y[i]) for i in range(num_point)]

total_area = 1**2
circle_area_estimated = total_area * sum(in_circle)/num_point

plt.contourf(X,Y,Z,0, colors=['lightblue', 'w'])
plt.scatter(x[in_circle], y[in_circle], color='r', s=10)
plt.scatter(x[~np.array(in_circle)], y[~np.array(in_circle)], color='g', s=10)
plt.title(f'Circle Shape \n Estimated pi value: {circle_area_estimated*4:.10f}')
plt.show()
```
