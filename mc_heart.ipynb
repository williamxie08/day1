```python
import matplotlib.pyplot as plt
import numpy as np

def is_in_heart(x,y):
    return x**2 + (5*y/4 - np.sqrt(np.abs(x)))**2 <= 1

num_point = 30000

plt.figure(figsize=(10,10))
x = np.linspace(-1.25, 1.25, 100)
y = np.linspace(-1, 1.5, 100)
X, Y = np.meshgrid(x,y)
Z = X**2 + (5*Y/4 - np.sqrt(np.abs(X)))**2 -1

plt.contourf(X,Y,Z,0, colors=['blue','w'])
plt.title('Circle shape')
plt.axis('equal')
plt.tight_layout()
plt.show()

plt.figure(figsize=(10,10))
np.random.seed(0)
x = np.random.uniform(-1.25, 1.25,num_point)
y = np.random.uniform(-1, 1.5,num_point)

in_heart = [is_in_heart(x[i],y[i]) for i in range(num_point)]

total_area = 2.5 * 2.5
heart_area_estimated = total_area * sum(in_heart)/num_point

plt.contourf(X,Y,Z,0, colors=['lightblue', 'w'])
plt.scatter(x[in_heart], y[in_heart], color='r', s=10)
plt.scatter(x[~np.array(in_heart)], y[~np.array(in_heart)], color='g', s=10)
plt.title(f'Circle Shape \n Estimated Area: {heart_area_estimated:.10f}')
plt.show()
```
