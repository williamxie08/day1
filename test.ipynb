```python
import numpy as np

NUMBER_TABLE = np.array([27, 16, 12, 25,  5, 17, 26, 23, 29,  9,  3, 20, 18, 19, 24,  8, 13,  2, 11, 15,  7,  6, 14, 22, 28,  4,  1, 10, 21])
np.random.shuffle(NUMBER_TABLE)
print(NUMBER_TABLE)

X_N = 0

def roll_dice():
    """randomly generate a number from 1 to 6, with a fixed sequence"""
    global X_N
    X_N = (X_N * 1664525 + 1013904223) % 2**32
    return int(X_N / 2**32 * 6) +1

for _ in range(20):
    while True:
        player_a = roll_dice()
        player_b = roll_dice()

        print(f'player A: {player_a}, Player B: {player_b}', end=', ')

        # if draw, roll again
        if player_a != player_b:
            break
        else:
            print ('Draw, roll again')

    winner = 'Player A' if player_a > player_b else "Player B"
    print (f'The winner is {winner}')

```
